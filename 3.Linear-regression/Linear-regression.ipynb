```python
import pandas as pd
import numpy as np
train = pd.read_csv('train.csv')
test = pd.read_csv('test_x.csv')
test = test.drop(test.columns[[0]],axis = 1)
train = train.drop(train.columns[[0]],axis = 1)

Y_train = train['rating_org']
train = train.drop(columns = ['rating_org'],axis = 1)
```

```python
#заполнить средним значением average_bill
av_train = train.loc[train['user_city']=='msk','average_bill']
av_train = av_train.fillna(av_train.mean())
train.loc[train['user_city']=='msk','average_bill'] = av_train

av_train = train.loc[train['user_city']=='spb','average_bill']
av_train = av_train.fillna(av_train.mean())
train.loc[train['user_city']=='spb','average_bill'] = av_train
del av_train

av_test = test.loc[test['user_city']=='msk','average_bill']
av_test = av_test.fillna(av_test.mean())
test.loc[test['user_city']=='msk','average_bill'] = av_test

av_test = test.loc[test['user_city']=='spb','average_bill']
av_test = av_test.fillna(av_test.mean())
test.loc[test['user_city']=='spb','average_bill'] = av_test
del av_test
test
```

```python
#Удаление столбцов
train = train.drop(columns = ['org_id'],axis = 1)
train = train.drop(columns = ['user_city'],axis = 1)
train = train.drop(columns = ['org_city'],axis = 1)

test = test.drop(columns = ['org_id'],axis = 1)
test = test.drop(columns = ['user_city'],axis = 1)
test = test.drop(columns = ['org_city'],axis = 1)
```

```python
#Работа с числовыми столбцами для train
train_numeric = train[['rating','ts','average_bill']]
train_numeric_cols = train_numeric.columns

from sklearn.preprocessing import StandardScaler
from scipy import stats
ss = StandardScaler()
train_numeric = pd.DataFrame(ss.fit_transform(train_numeric))
train_numeric.columns = train_numeric_cols


#Работа с числовыми столбцами для test
test_numeric = test[['rating','ts','average_bill']]
test_numeric_cols = test_numeric.columns
test_numeric = pd.DataFrame(ss.fit_transform(test_numeric))
test_numeric.columns = test_numeric_cols
test_numeric
```

```python
#Работа с bool и категориями для train
train_cat = train.drop(['rating','ts','average_bill'], axis = 1)
train_cat = train_cat.merge(pd.get_dummies(train_cat.rubrics, drop_first=True), left_index=True, right_index=True)
train_cat.drop('rubrics', axis = 1, inplace=True)

#Работа с bool и категориями для test
test_cat = test.drop(['rating','ts','average_bill'], axis = 1)
test_cat = test_cat.merge(pd.get_dummies(test_cat.rubrics, drop_first=True), left_index=True, right_index=True)
test_cat.drop('rubrics', axis = 1, inplace=True)
test_cat
```

```python
#Слить таблицы
train = pd.merge(train_numeric, train_cat, left_index = True, right_index = True)
test = pd.merge( test_numeric, test_cat, left_index = True, right_index = True)
test
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
X_reg=train
y_reg=Y_train

kb = SelectKBest(score_func=f_regression, k=50)
kb.fit(X_reg, y_reg)
features_list = X_reg.columns
chosen_list = features_list[kb.get_support()]
print(features_list[kb.get_support()])
```

```python
#Превращаем в массивы
y_train = Y_train.values
train = train[chosen_list]
x_train = train.values
test = test[chosen_list]
x_test = test.values


# Делим на тренировочную и тестовую
from sklearn.model_selection import train_test_split
x_t1,x_t2,y_t1,y_t2 = train_test_split(x_train, y_train, test_size=0.7, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
ml = LinearRegression()
ml.fit(x_t1, y_t1)
y_pred=ml.predict(x_t2)
y_pred_test=ml.predict(x_test)
print(y_pred)
print(y_pred_test)
```

```python
#оценка погрешности
import numpy as np
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
print(mean_absolute_percentage_error(y_t2,y_pred))
from sklearn.metrics import r2_score
print(r2_score(y_t2,y_pred))
```
